```python
from langchain.text_splitter import(
    RecursiveCharacterTextSplitter,
    Language,
)
```

```python
RecursiveCharacterTextSplitter.get_separators_for_language(Language.PYTHON)
```

```python
PYTHON_CODE = """
def hello_world();
    print("Hello, World!")
hello_world()
"""

python_splitter = RecursiveCharacterTextSplitter.from_language(
    language=Language.PYTHON, chunk_size=50, chunk_overlap=0
)
python_docs = python_splitter.create_documents([PYTHON_CODE])
python_docs
```

```python
!pip install tiktoken

import tiktoken
tokenizer = tiktoken.get_encoding("cl100k_base")

def tiktoken_len(text):
  tokens = tokenizer.encode(text)
  return len(tokens)
```

```python
tiktoken_len(texts[1].page_content)
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000, chunk_overlap=0, length_function=tiktoken_len
)
texts = text_splitter.split_documents(pages)
```

```python
print(len(texts[1].page_content))
print(tiktoken_len(texts[1].page_content))
```

```python
text_splitter = RecursiveCharacterTextSplitter.from_tiktoken_encoder(
    chunk_size=1000, chunk_overlap=0
)
texts = text_splitter.transform_documents(pages)
```

```python
print(len(texts[1].page_content))
print(tiktoken_len(texts[1].page_content))
```

```python
token_list = []
for i in range(len(texts)):
  char_list.append(tiktoken_len(texts[i].page_content))
print(char_list)
```
